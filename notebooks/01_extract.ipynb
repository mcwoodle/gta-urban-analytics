## Verify Downloads

List the files that landed in `data/01_raw/`:

```python
import os

raw_dir = os.path.join('..', 'data', '01_raw')
for f in sorted(os.listdir(raw_dir)):
    size_mb = os.path.getsize(os.path.join(raw_dir, f)) / (1024 * 1024)
    print(f'{f:50s} {size_mb:8.2f} MB')
```
